## Clean column names

```python
#drop the historical column 
def clean_columns(df):
    df.columns = [column.replace(" ", "_").lower() for column in df.columns]
```

```python
clean_columns(ev)
```

```python
ev
```

## Remove the column category 

```python
def drop_category (df):
    del df['category']
```

```python
drop_category(ev)
```

```python
ev
```

## Remove world from region column 

```python
#drop the world row
def drop_world(df):
    df.drop(df[df['region'] == 'World'].index, inplace=True)
```

```python
drop_world (ev)
```

```python
ev
```

## Keep only EV sales from the parameter column.

```python
def keep_ev_sales(df):
    return df[df['parameter'] == 'EV sales']
```

```python
keep_ev_sales(ev)
```

## Keep relevant columns

```python
ev_filtered = ev[["region","year", "value"]]
```
